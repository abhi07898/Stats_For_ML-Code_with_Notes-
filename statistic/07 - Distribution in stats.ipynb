```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
n,p = 1,.5  #number of trails, probability of each trail 
s = np.random.binomial(n, p, 1000)

plt.hist(s)
```

```python
from scipy.stats import bernoulli
import seaborn as sns

bernoulli_data = bernoulli.rvs(size=1000, p=0.6)
ax = sns.distplot(bernoulli_data,
                kde = True,
                color = 'crimson', 
                hist_kws = {'linewidth':25, 'alpha':1})
ax.set(xlabel = 'Bernouli', ylabel = 'Frequency')
```

# 2. Geometry Distribution 

```python
from __future__ import print_function, division 
from scipy.stats import geom
```

```python
prob = 0.2
n = 12
p = np.zeros(n)
q = np.zeros(n)
for k in range(1,n+1):
    p[k-1] = (1-prob)**(k-1)*prob
    q[k-1] = geom.pmf(k,prob)
plt.bar(range(1,13), p, color = 'green')
plt.xlim(1,15)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.title('first success at n ')
```

# 3. Binomial Distribution 

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10, p=0.5, size=10000)
ax = sns.distplot(data_binom,
                 kde = False,
                 color = 'orange',
                 hist_kws = {"linewidth":15,'alpha':1})
ax.set(xlabel='Binomial', ylabel='frequency')
```
